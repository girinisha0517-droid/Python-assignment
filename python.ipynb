# 12. How does the elif statement work
Ans - The elif statement , short for "else if" , allows for the evaluation of multiple conditions . It allows for a more stuctured and efficient way to handle conditional logic when there are several possible outcomes based on different critria.

HOW IT Works:

   . Initial if check :  The program first evaluates the condition in the if statement .If this condition is True , the code block associated with the if statement is executed , and the entire if-elif-else block is exited .
   
   .Sequential elif checks : If the if condition is False , the program then moves to the first elif statement and evaluates its condition . If this elif condition is true , its associate code block is executed ,and the entire if-elif-else block is exited.

   .Optional else block : if none of the if or elif conditions evaluate to True , the code block within the optional else statement (if present ) is executed .
